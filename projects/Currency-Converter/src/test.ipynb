```python
from cachetools import cached , TTLCache
```

```python
ttlcache = TTLCache(maxsize= 100 , ttl=5)
```

```python
@cached(ttlcache)
def fibo(n):
    if n < 2:
        return n
    return fibo(n-1) + fibo (n -2)
```
