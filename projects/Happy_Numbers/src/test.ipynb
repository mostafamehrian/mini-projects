```python
def is_happy(n):
    seen = set()
    while (n != 1) and (n not in seen):
        seen.add(n)
        n = sum([int(i) **2 for i in str(n)])
        
    return n == 1
```

```python
get_number = int(input("enter a number for checking is number is happ or not? "))
is_happy(get_number)
```
